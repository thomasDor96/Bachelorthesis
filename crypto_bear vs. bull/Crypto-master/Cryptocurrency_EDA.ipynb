```python
import pandas as pd
import numpy as np
import datetime
pd.options.display.max_rows = 100
pd.set_option('display.float_format', lambda x: '%.8f' % x)
```

```python
crypto_markets = pd.read_csv("/Users/shubhvarma/Downloads/crypto-markets.csv")
```

```python
crypto_markets = crypto_markets.assign(date = pd.to_datetime(crypto_markets['date']))
```

```python
crypto_markets[crypto_markets['symbol'] == 'BAT']
```

```python
split_date = '2017-01-01'
markets17 = crypto_markets[crypto_markets['date'] >= split_date]
```

```python
july_price = markets17[markets17.date == '2017-07-01'].loc[:,['slug','symbol','open','market','volume']]
oct_price = markets17[markets17.date == '2017-10-1'].loc[:,['slug','symbol','open','market','volume']]
dec_price = markets17[markets17.date == '2017-12-31'].loc[:,['slug','symbol','open','market','volume']]
halfYear_priceChange = pd.merge(july_price, oct_price, how='outer', on='slug',suffixes=('_july1',''))
priceChange = pd.merge(halfYear_priceChange, dec_price, how='outer',on='slug',suffixes=('_oct1','_dec31'))
```

```python
priceChange['Q3_change'] = priceChange['open_oct1']/priceChange['open_july1'] - 1
priceChange['Q4_change'] = priceChange['open_dec31']/priceChange['open_oct1'] - 1
priceChange['halfYear_change'] = priceChange['open_dec31']/priceChange['open_july1'] - 1
orderCols = ['slug','symbol_july1','halfYear_change','Q4_change','Q3_change','market_dec31','market_oct1','market_july1','volume_dec31','volume_oct1','volume_july1','open_dec31','open_oct1','open_july1']
priceChange.sort_values(['Q4_change'],ascending=False)[orderCols].reset_index().drop(['index'],axis=1)[36:]
```
