```python
#Json file with ALL cryptocurrencies
import pandas as pd

crypto_all = pd.read_json('https://api.coinmarketcap.com/v1/ticker/?limit=0')
crypto_all
```

```python
crypto_all.shape
```

```python
crypto_all.keys()
```

```python
market_cap_raw = crypto_all[['id', 'market_cap_usd']]
market_cap_raw
```

```python
market_cap_raw.describe()
```

```python
print('Number of currencies: ' + str(len(market_cap_raw)))
print('Number of currencies with market capitalization: {}'.format(market_cap_raw.market_cap_usd.count()))
```

```python
#pd.query() method uses numexpr tech., logical/boolean operators must be in &&, !, | format
market_cap_clean = market_cap_raw.query('market_cap_usd > 0')
#Alternative for ^: market_cap_clean = market_cap_raw[market_cap_raw.market_cap_usd > 0]

print('Number of currencies: ' + str(len(market_cap_clean)))
print('Number of currencies with market capitalization: {}'.format(market_cap_clean.market_cap_usd.count()))
market_cap_clean
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg' 
plt.style.use('fivethirtyeight')

#Declaration of str variables for plot titles
TOP_CAP_TITLE = 'Top 10 market capitalization'
TOP_CAP_Y = '% of Total Cap'
TOP_CAP_X = 'Cryptocurrency'

cap10 = market_cap_clean[0:10]
average_market_cap = market_cap_clean.market_cap_usd.sum()/len(market_cap_clean)
total_market_cap = market_cap_clean.market_cap_usd.sum()
print('The average market Capitalization of all currencies with market cap.: ${:,.2f}'.format(average_market_cap))
print('Total market capitalization: ${:,.2f}'.format(total_market_cap))


cap10 = cap10.assign(market_cap_per = lambda x: (x.market_cap_usd/total_market_cap)*100)
```

```python
cap10
```

```python
plt.barh(cap10.id, cap10.market_cap_per)
plt.show()
```
